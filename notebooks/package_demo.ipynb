### PACKAGE NAME : CQM-solver for VRP using D-Wave's Quantum Annealing hardware and SDK

#### Getting started

`PACKAGE NAME` is a combinatorial solver for VRP using D-Wave's Quantum Annealing hardware. Currently, the package provides two class of solvers, namely the TSP and ESP solvers. The TSP solver finds the shortest hamiltonian cycle for a given graph, and the ESP solver finds the shortest path between two nodes. 

We start by importing the package.

```python
from cqm.solver import QTSP_Improved, QESP
```

We will first run the `QESP` solver.

```python
# Define the graph
coordinates = [[1, 1], [2, 3], [3, 2], [2, 4], [1, 5], [3, 6]]
edges = [[0, 1], [1, 2], [1, 3], [1, 5], [2, 3], [3, 4], [4, 5]]

# Define the solver
esp = QESP(coordinates=coordinates, edges=edges, source=1, destination=4, time=30, log=False)

# Run the solver
esp(token=token)
```
